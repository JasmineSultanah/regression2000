```python
# imports
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_table("clds.txt")
```

```python
data.head(5)
```

```python
for col in data.columns[2:5]:
    data[col] = pd.to_datetime(data[col],infer_datetime_format=True)
```

```python
data.head(5)
```

```python
data = data.sort_values(by='JoinedGroup')
```

```python
data.head()
```

```python
dates = pd.DataFrame({"Count":pd.Series(pd.value_counts(data.JoinedGroup)),
                      "Date":pd.Series(pd.value_counts(data.JoinedGroup)).index}
                    )
```

```python
dates = dates.sort_values(by='Date')
sns.tsplot(data=dates,time="Date",value="Count")
plt.show()
```
